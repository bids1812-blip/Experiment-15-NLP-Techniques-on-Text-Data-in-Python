```python
#Word Tokenization
from nltk.tokenize import word_tokenize
text = "Natural Language Processing is interesting"
tokens = word_tokenize(text)
print(tokens)
```

```python
#Sentence Tokenization
from nltk.tokenize import sent_tokenize
text = "Python is easy. It is widely used in data science."
sentences =  sent_tokenize(text)
print(sentences)
```

```python
#Stop word removal
#Stop words such as is, the, and, in are removed
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))    #Don't write 'English'
words = word_tokenize(text)
filtered_words = [w for w in words if w.lower() not in stop_words]
print(filtered_words)
```

```python
#Stemming
#Stemming reduces words to their root form

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["playing", "running", "typing", "writing", "swimming", "studies"]

for w in words:
  print(stemmer.stem(w))
```

```python
#Lemmatization
#Lemmatization converts words to their dictionary form

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("playing"))
print(lemmatizer.lemmatize("writing"))
print(lemmatizer.lemmatize("swimming"))
print(lemmatizer.lemmatize("studies"))
print(lemmatizer.lemmatize("better"))
```

```python
#Part of Speech (POS) Tagging
#POS tagging identifies grammatical roles of words
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

```

```python
from nltk import pos_tag
words = word_tokenize("Python is a powerfu; programming language")
pos_tags = pos_tag(words)
print(pos_tags)
```
